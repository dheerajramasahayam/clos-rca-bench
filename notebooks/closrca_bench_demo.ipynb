# ClosRCA-Bench End-to-End Demo

This notebook walks through the public research artifact at a reviewer-friendly level: load the benchmark snapshot, inspect leaderboard outputs, and preview the main figure assets without rerunning the full training pipeline.


## Setup

Run this notebook from the repository root or from `notebooks/`.


````python
from pathlib import Path
import json
import pandas as pd
from IPython.display import Image, display

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

ROOT
````

## Load the benchmark snapshot


````python
snapshot_path = ROOT / "examples" / "closrca_bench_snapshot.json"
with snapshot_path.open("r", encoding="utf-8") as handle:
    snapshot = json.load(handle)

snapshot
````
